```r
library(ggplot2)
library(viridis)
library(dplyr)
```

```r
x <- rnorm(10000)
y <- rnorm(10000)
```

```r
data <- data.frame(x =x, y = y)
```

```r
ggplot(data, aes(x = x, y = y)) + geom_density_2d() + theme_bw()
```

```r
ggplot(data, aes(x = x, y = y)) + geom_bin_2d() + theme_bw()
```

```r
ggplot(data, aes(x = x, y = y)) + geom_bin_2d(bins = 70)  + scale_fill_viridis_b() + theme_bw()
```

```r
p <- ggplot(data, aes(x = x, y = y)) + geom_hex(bins = 50)  + scale_fill_viridis_b(option = "inferno") + theme_bw()
p
```

```r
png("imagenes/hexbin.png")
p
dev.off()
```

```r
## Vamos a revisar otro grafico, el de burbujas, basicamente es un scatrerplot 
## pero añadimos una 3ra varible que se vera reflejada en el tamaño de los puntos
```

```r
pkm <- read.csv("data/pkm.csv")
```

```r
head(pkm)
```

```r
pkm2 <- pkm |> filter(Generation == 1)

q <- pkm2 |> ggplot( aes(x = Attack, y = Sp..Atk, size = HP, fill = Speed)) + geom_point(alpha = 0.6, shape = 21, color = "black") + 
    scale_size(range = c(.1, 20), name="Salud") + scale_fill_viridis_c(option = "magma") +
    theme_bw() + xlab("Ataque") + ylab("Ataque Especial") + ggtitle("Relacion entre ataque y ataque especial") 
q
```

```r
png("imagenes/burbujas.png")
q
dev.off()
```
