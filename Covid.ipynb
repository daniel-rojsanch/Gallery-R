### Covid México

```r
Casos <- read.csv("data/CasosCovid.csv")
Defun <- read.csv("data/Defunciones.csv")
```

```r
tail(Defun)
```

```r
Casos$nombre
```

```r
inicio <- as.Date("2020-02-26")
fin    <- as.Date("2022-05-22")
dias   <- seq(inicio, fin, by = 1)
```

```r
iniciodefun <- as.Date("2020-03-17")
findefun    <- as.Date("2022-05-20")
dias2 <- seq(iniciodefun, findefun, by = 1)
```

```r
# png("imagenes/covidEdoMex.png", width = 700, height = 500)

plot(dias, Casos[Casos$nombre == 'MEXICO',4:820], type = 'l', 
     main = "Casos confirmados de Covid \n Estado de México ", ylab = "Casos diarios", 
     col = "blue")

lines(dias2, Defun[Defun$nombre == 'MEXICO',4:798], type = 'l', 
     col = "red", lwd = 2)

legend("left", legend=c("Muertes","Contagios"), col=c("red", "blue"), pt.cex=2, pch=15 )

grid()

# dev.off()
```

```r
x <- as.numeric( Casos[Casos$nombre == 'MEXICO',26:820] )
y <- as.numeric( Defun[Defun$nombre == 'MEXICO',4:798] )
```

```r
length(x)
length(y)
```

```r
plot(x, y, main = "Contagios vs Defunciones", sub = "Estado de México",
    xlab = "Casos Confirmados", ylab = "Defunciones", col = rgb(.7 , .2 , .4, .3), lwd = 4)
grid()
```
